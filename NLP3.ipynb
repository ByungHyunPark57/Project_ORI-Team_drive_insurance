```python
## 어퍼스트로피 풀기

# 수축형 대체 패턴
change = {
    r"\'m": " am",
    r"\'s": " is",
    r"\'ve": " have",
    r"n\'t": " not",
    r"\'re": " are",
    r"\'d": " had",
    r"\'ll": " will",
}

# 수축형 대체 함수
def replace_contractions(text):
    for pattern, replacement in change.items():
        text = re.sub(pattern, replacement, text)
    return text
```

```python
train['text'] = train['text'].apply(replace_contractions)
test['text'] = test['text'].apply(replace_contractions)
```

```python
# 첫 글자를 소문자로 변환하는 함수
def lower_first_letter(text):
    if text:
        return text[0].lower() + text[1:]
    return text
```

```python
# 텍스트 전처리 함수

def text_preprocessing(text_list):

    stop_words = set(stopwords.words('english'))
    token_list = []

    for text in text_list:
        txt = re.sub('[^A-Za-z0-9 ]', ' ', text)  #영어 대,소문자만 남기고 다른 글자 제거
        tokens = word_tokenize(txt)
        tokens = [t for t in tokens if t.lower() not in stop_words]
        token_list.append(tokens)

    return token_list
```

```python
# 데이터 전처리

train['text'] = train['text'].apply(lower_first_letter)
train['text'] = text_preprocessing(train['text']) #stop_words 제거 때문에 2번 해야 함??


test['text'] = test['text'].apply(lower_first_letter)
test['text'] = text_preprocessing(test['text'])
```
